### Helper functions

```python
def get_target_distribution(df, feature_name):
    """Takes in dataframe, returns distribution of target depending on values"""
    df_tmp = df.copy()
    df_tmp[feature_name] = df_tmp[feature_name].fillna('na')
    return df_tmp.groupby(feature_name)['target'].agg('mean')
```
